# MATPLOT

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x=np.linspace(0, np.pi*4, num=200)

x
```

```python
y=np.sin(x)
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('Sin (x)')
plt.show()
```

```python
z=np.cos(x)

plt.plot(x,y,'g', label='sin(x)')
plt.plot(x,z, 'b', label='cos(x)')
plt.xlabel('x')
plt.xlim([0,6])
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,z,'b')

#plt.tight_layout()
plt.show()
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y,label='sin(x)')
axes.legend()
plt.show()
```

```python
import pandas as pd
```

```python
col_names= ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car']
```

```python
df=pd.read_csv('auto-mpg.data', sep='\s+', names=col_names)
df
```

```python
plt.hist(df['mpg'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['mpg'], bins = 20)
plt.xlabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=df['weight'], y=df['mpg'], alpha=0.5)
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist2d(x=df['weight'], y=df['mpg'], alpha=0.5)
plt.xlabel('weight')
plt.ylabel('mpg')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hexbin(x=df['weight'], y=df['mpg'], alpha=0.5, gridsize=(35,15), cmap='jet')
plt.xlabel('weight')
plt.ylabel('mpg')
plt.colorbar()
plt.show()
```

# Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df.head()
```

```python
sns.distplot(df['mpg'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.distplot(df['mpg'], kde=False, bins=20)
plt.show()
```

```python
sns.lmplot('weight','mpg', data=df)
plt.show()
```

```python
sns.lmplot('weight','mpg', data=df, fit_reg= False, aspect=2, size=4)
plt.show()
```

```python
sns.lmplot('weight','mpg', data=df, fit_reg= False, aspect=2, size=4, hue='cylinders')
plt.show()
```

```python
sns.lmplot('weight','mpg', data=df, fit_reg= False, aspect=1, size=5, hue='cylinders', col='origin')
plt.show()
```

```python
sns.kdeplot(df['weight'],df['mpg'])
plt.show()
```

```python
sns.kdeplot(df['weight'],df['mpg'], shade=True)
plt.show()
```

```python
sns.jointplot('weight', 'mpg', data=df, alpha=0.4)
```

```python
g=sns.PairGrid(df, diag_sharey=False)
g.map_lower(sns.kdeplot)
g.map_upper(plt.scatter)
g.map_diag(sns.kdeplot)
```

```python
sns.countplot(y='cylinders', data=df)
```

```python
sns.boxplot(x='origin', y='mpg', data=df, hue='cylinders')
```

# Graph with pandas

```python
df.head()
```

```python
df['weight'].plot(kind='hist', bins=20)
```

```python
df.groupby('cylinders')['mpg'].plot(kind='hist', alpha=0.4, legend=True)
```

```python
df.plot(kind='scatter', x='weight', y='mpg')
plt.show()
```

```python
df.boxplot(column='mpg', by='cylinders', grid=False)
plt.show()
```

```python
dfp=df.pivot_table(index='cylinders', aggfunc='mean')
dfp
```

```python
dfp.plot.line(y='acceleration')
```

```python
for col in dfp.columns:
    dfp.plot.line(y=col)
```

```python
df.groupby('cylinders').count()['mpg'].plot(kind='bar')
plt.show()
```
